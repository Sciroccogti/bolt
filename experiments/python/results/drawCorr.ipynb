```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt

def plotCorr(X: np.ndarray):
    corr = np.corrcoef(X[:256,:])
    abs_matrix = np.abs(corr)
    plt.imshow(abs_matrix)
    plt.colorbar()
```

```python
%matplotlib widget
fig1 = plt.figure()
X1 = np.load("../../assets/dft-rms500-10dB/IDFT_X.npy")
corr1 = np.corrcoef(X1[:256,:])
abs_matrix1 = np.abs(corr1)
plt.imshow(abs_matrix1)
plt.colorbar()
plt.title("IDFT")
plt.show()
plt.savefig("zIDFT.svg")

fig2 = plt.figure()
X = np.load("../../assets/dft-rms500-10dB/DFT_X.npy")
corr = np.corrcoef(X[:256,:])
abs_matrix = np.abs(corr)
plt.imshow(abs_matrix)
plt.colorbar()
plt.title("DFT")
plt.show()
plt.savefig("zDFT.svg")
```
